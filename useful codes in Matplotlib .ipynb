# First graph

First we need are sure with your codes run in python 2 and 3, for this reason import the next libraries

```python
from __future__ import division, print_function, unicode_literals
```

For the graphics we need the matplotlib package and we call matplotlib with **plt** to all functions and adittional magic command **%matplotlib inline** for show images in notebook cells. 

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Now, let's go to the first graph. For this example we consider a mathematical funcion and the NumPy package help for this work. The general expresion it's

$$f(x) = Ax^2 + Bx + C$$

We use the Numpy's linspace function to create the domain $x$ of the function an array with 50 float numbers in the range $[-2,2]$.

```python
x = np.linspace(-2,2,50)
```

Next we make the mathematical function with the next syntax

```python
def f(x,A,B,C):
    return A*x**2+B*x+C
```

the following instruccion plot the previous function $f$. In this case 

$A = 1\qquad B=2 \qquad C = 3$

```python
plt.plot(x,f(x,1,2,3))
```

This plot is very simple, we added a title, x and y labels and a grid with the next commands

```python
plt.plot(x,f(x,1,2,3))
plt.title("Square function")
plt.xlabel("$x$ [domain]")
plt.ylabel("$f(x)$ [Range]")
plt.grid(True)
```

If the plot needs a label , the label command help. The $$ make a Latex format

```python
plt.plot(x,f(x,1,2,3), label = "$f(x) = x^2 + 2x + 3$")
plt.title("Square function")
plt.xlabel("$x$ [domain]")
plt.ylabel("$f(x)$ [Range]")
plt.legend()
plt.grid(True)

```

## Styles and colors

```python
plt.plot(x,f(x,1,2,3), "g--",label = "$f(x) = x^2 + 2x + 3$")
plt.plot(x,f(x,5,6,7), "b:",label = "$f(x) = 5x^2 + 6x + 7$")
plt.plot(x,f(x,1,8,0), "r^",label = "$f(x) = x^2 + 8x $")
plt.plot(x,f(x,5,0,0), "k+",label = "$f(x) = 5x^2 $")

plt.title("Square function")
plt.xlabel("$x$ [domain]")
plt.ylabel("$f(x)$ [Range]")
plt.legend()
plt.grid(True)

```

Matplotlib contains subplots organized in grids, for make this we call the subplot function and specify the number of rows and columns in the figure 

```python
plt.subplot(2,2,1)
plt.plot(x,f(x,1,2,3), "g--",label = "$f(x) = x^2 + 2x + 3$")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,f(x,5,6,7), "b:",label = "$f(x) = 5x^2 + 6x + 7$")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x,f(x,1,8,0), "r-",label = "$f(x) = x^2 + 8x $")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,f(x,5,0,0), "k+",label = "$f(x) = 5x^2 $")
plt.grid(True)
```
